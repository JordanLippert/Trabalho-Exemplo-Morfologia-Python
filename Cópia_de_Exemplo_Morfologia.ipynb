```python
#from https://opencv-python-tutroals.readthedocs.io/en/latest/py_tutorials/py_imgproc/py_morphological_ops/py_morphological_ops.html
import cv2
import numpy as np
#caso for usar o Google Colab com a OpenCV, usar a lib abaixo
from google.colab.patches import cv2_imshow
```

```python
img = cv2.imread('j.png',0) #Nesse começo, na 1ª linha, está sendo lido a imagem "j.png" em escala de cinza usando a função "cv2.imread()". A imagem daí depois é armazenada na variável "img".
img_opening = cv2.imread('j_ruido.png',0) #Nessa 2ª linha, está sendo lida outra imagem, a "j_ruido.png", também em escala de cinza usando a função  "cv2.imread()". E no final é armazenada em outra variável chamada "img_opening".
img_closing = cv2.imread('j_furos.png',0) #Na 3ª linha, está sendo lida uma 3ª imagem chamada "j_furos.png", novamente lida em escala de cinza, e sendo armazenada na variável "img_closing".
altura, largura = img.shape #Aqui, utilizamos o "Método Shape" na variável "img" para obtermos as dimensões da imagem. O "Método Shape" é uma função disponível em muitas bibliotecas de processamento de imagem, como o OpenCV em Python. Quando o usuamos esse método, nos será retornado as dimensões de uma matriz ou de uma imagem, como é o nosso caso. A altura será armazenada na variável "altura", e a largura será armazenada na váriavel "largura".
kernel = np.ones((5,5),np.uint8) #Na penúltima linha, estamos criando um kernel, que é uma matriz de valores 1, usando a função "np.ones()". O kernel é uma matriz 5x5 preenchida com uns e será do tipo "np.uint8". Esse kernel daí será usado em operações morfológicas, como abertura e fechamento.
print(kernel) #E por fim, estamos imprimindo o kernel na saída padrão para podermos visualizar a sua estrutura. Isso nos ajuda a entender como a matriz 5x5 de uns é formada.
```

```python
erosion = cv2.erode(img,kernel,iterations = 2) #Nesse início, está sendo realizado a operação de erosão na imagem "img", usando a função "cv2.erode()". A erosão é uma operação morfológica que reduz o tamanho dos objetos presentes na imagem. O 1º argumento nos parênteses "img", representa a imagem de entrada na qual a erosão será aplicada. O 2º argumento nos parênteses "kernel", é o kernel que define o padrão da erosão. E o 3º argumento nos parênteses "iterations=2", especificará o número de vezes que a erosão será aplicada. No nosso caso, a erosão será aplicada 2x consecutivas. O resultado da erosão é armazenado daí na variável "erosion".
dilation = cv2.dilate(img,kernel,iterations = 2) #Aqui na 2ª linha, está sendo realizado a operação de dilatação na imagem "img", usando a função "cv2.dilate()". A dilatação é uma outra operação morfológica que aumenta o tamanho dos objetos presentes na imagem. Assim como antes, o 1º argumento nos parênteses é "img", é a imagem de entrada na qual a dilatação será aplicada. O 2º argumento nos parênteses "kernel", é o kernel utilizado para realizar a dilatação. E por fim, o 3º argumento nos parênteses "iterations=2", definirá o número de vezes que a dilatação será aplicada. Novamente como antes, a dilatação será aplicada 2x consecutivas. O resultado da dilatação daí é armazenado na variável "dilation".
#As operações de erosão e dilatação são amplamente utilizadas no processamento de imagens para realizar tarefas como a remoção de ruídos, preenchimento de lacunas em objetos e alteração de formas. E são particularmente úteis em operações de segmentação de imagens e análise de formas.
```

```python
gradient = cv2.morphologyEx(img, cv2.MORPH_GRADIENT, kernel) #Aqui no começo, está sendo realizada uma operação de gradiente morfológico na imagem "img" usando a função "cv2.morphologyEx()". O gradiente morfológico é uma operação que destaca as bordas da imagem ao calcular a diferença entre a dilatação e a erosão da imagem. O 1º argumento nos parênteses "img" é a imagem de entrada na qual o gradiente morfológico será aplicado. O 2º argumento nos parênteses "cv2.MORPH_GRADIENT" especifica o tipo de operação morfológica a ser realizada, que no nosso caso é o gradiente. O 3º argumento nos parênteses "kernel", é o kernel utilizado para a operação morfológica. O resultado do gradiente morfológico será armazenado daí na variável "gradient".
opening = cv2.morphologyEx(img_opening, cv2.MORPH_OPEN, kernel) #Na 2ª linha, está sendo realizada uma operação de abertura morfológica na imagem "img_opening" usando a função "cv2.morphologyEx()". A abertura morfológica é uma operação que suaviza as bordas e remove pequenos detalhes de uma imagem. O 1º argumento nos parênteses é o "img_opening", que é a imagem de entrada na qual a abertura será aplicada. O 2º argumento nos parênteses "cv2.MORPH_OPEN" indica que queremos realizar uma operação de abertura. O 3º argumento nos parênteses "kernel", é o kernel utilizado para a operação morfológica. O resultado da abertura morfológica será armazenado daí na variável "opening".
closing = cv2.morphologyEx(img_closing, cv2.MORPH_CLOSE, kernel) #E por fim, nessa última linha, está sendo realizada uma operação de fechamento morfológico na imagem "img_closing" usando a função "cv2.morphologyEx()". O fechamento morfológico é uma operação que preenche lacunas em objetos e une regiões próximas. O 1º argumento nos parênteses é o "img_closing" é a imagem de entrada na qual o fechamento será aplicado. O 2º argumento nos parênteses é o "cv2.MORPH_CLOSE" que indica que queremos realizar uma operação de fechamento. O 3º argumento nos parênteses é o "kernel", esse kernel é utilizado para a operação morfológica. O resultado do fechamento morfológico é armazenado então na variável "closing".
#Essas operações morfológicas são amplamente utilizadas em processamento de imagens para realizar transformações e manipulações em objetos e regiões de interesse em uma imagem.
```
